```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
  return TextBlob(str(text)).sentiment.polarity
  # blob=TextBlob(str(text)).sentiment.polarity
  # if blob > 0:
  #   return "positive"
  # elif blob < 0:
  #   return "negative"
  # else:
  #   return "neutral"
```

```python
df["Sentiment"]=df["fullText"].apply(get_sentiment)
```

```python
ax1=plt.gca()
ax2=ax1.twinx()

plt.title('The Musk Effect: Sentiment vs Reality')

ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Return (%)', color="red")
ax1.plot(final_df.index, final_df['Stock_Return'], color="red", label='Stock Return')
ax1.tick_params(axis='y', labelcolor="red")

ax2.set_ylabel('Musk Sentiment', color="blue")
ax2.plot(final_df.index, final_df['Sentiment'], color="blue", linestyle='--', label='Sentiment')
ax2.tick_params(axis='y', labelcolor="blue")

plt.show()
```

```python
subset = final_df.tail(30)

fig, ax1 = plt.subplots(figsize=(12,6))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Return (%)', color=color)
ax1.plot(subset.index, subset['Stock_Return'], color=color, label='Stock Return')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Musk Sentiment', color=color)
ax2.plot(subset.index, subset['Sentiment'], color=color, linestyle='--', label='Sentiment')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('The Musk Effect: Sentiment vs Reality')
plt.show()
```

```python
mean_mood = final_df['Sentiment'].mean()
std_mood = final_df['Sentiment'].std()
```

```python
print(f"Elon's Average Mood: {mean_mood:.4f}")
print(f"Typical Swing (Std Dev): {std_mood:.4f}")
```

```python
threshold = 1.5
upper_bound = mean_mood + (threshold * std_mood)
lower_bound = mean_mood - (threshold * std_mood)
```

```python
print(f"\nFiltering for days where Sentiment > {upper_bound:.3f} or < {lower_bound:.3f}...")
```

```python
extreme_df = final_df[
    (df['Sentiment'] > upper_bound) |
    (df['Sentiment'] < lower_bound)
]

print(f"Original Days: {len(final_df)}")
print(f"Extreme Days Found: {len(extreme_df)}")
```

```python
print("\n--- THE EXTREME TRUTH ---")

if len(extreme_df) > 10: # We need at least a few data points to correlate
    # Does EXTREME sentiment predict the next day?
    corr_extreme = extreme_df['Sentiment_lag_1'].corr(extreme_df['Stock_Return'])

    print(f"Correlation on Extreme Days: {corr_extreme:.4f}")

    if abs(corr_extreme) > 0.1:
        print(">> SIGNAL FOUND: When Elon screams, the market listens.")
    else:
        print(">> NO SIGNAL: Even when he screams, the market ignores him.")
else:
    print("Not enough extreme days to calculate correlation.")
```
